```python
day1=pd.read_csv('./data/서울시_강우량_정보_2021년01월.csv',encoding='cp949')
day2=pd.read_csv('./data/서울시_강우량_정보_2021년02월.csv',encoding='cp949')
day3=pd.read_csv('./data/서울시_강우량_정보_2021년03월.csv',encoding='cp949')
day4=pd.read_csv('./data/서울시_강우량_정보_2021년04월.csv',encoding='cp949')
day5=pd.read_csv('./data/서울시_강우량_정보_2021년05월.csv',encoding='cp949')
day6=pd.read_csv('./data/서울시_강우량_정보_2021년06월.csv',encoding='cp949')
day7=pd.read_csv('./data/서울시_강우량_정보_2021년07월.csv',encoding='cp949')
day8=pd.read_csv('./data/서울시_강우량_정보_2021년08월.csv',encoding='cp949')
day9=pd.read_csv('./data/서울시_강우량_정보_2021년09월.csv',encoding='cp949')
day10=pd.read_csv('./data/서울시_강우량_정보_2021년10월.csv',encoding='cp949')
day11=pd.read_csv('./data/서울시_강우량_정보_2021년11월.csv',encoding='cp949')
day12=pd.read_csv('./data/서울시_강우량_정보_2021년12월.csv',encoding='cp949')
```

```python
day1['자료수집 시각']= pd.to_datetime(day1['자료수집 시각']).dt.date
day2['자료수집 시각']= pd.to_datetime(day2['자료수집 시각']).dt.date
day3['자료수집 시각']= pd.to_datetime(day3['자료수집 시각']).dt.date
day4['자료수집 시각']= pd.to_datetime(day4['자료수집 시각']).dt.date
day5['자료수집 시각']= pd.to_datetime(day5['자료수집 시각']).dt.date
day6['자료수집 시각']= pd.to_datetime(day6['자료수집 시각']).dt.date
day7['자료수집 시각']= pd.to_datetime(day7['자료수집 시각']).dt.date
day8['자료수집 시각']= pd.to_datetime(day8['자료수집 시각']).dt.date
day9['자료수집 시각']= pd.to_datetime(day9['자료수집 시각']).dt.date
day10['자료수집 시각']= pd.to_datetime(day10['자료수집 시각']).dt.date
day11['자료수집 시각']= pd.to_datetime(day11['자료수집 시각']).dt.date
day12['자료수집 시각']= pd.to_datetime(day12['자료수집 시각']).dt.date
```

```python
df=pd.concat([day1,day2,day3,day4,day5,day6,day7,day8,day9,day10,day11,day12])
```

```python
df['자료수집 시각'] = pd.to_datetime(df['자료수집 시각'], format='%Y-%m-%d')
df.info()
```

```python
# 자료수집 시각, 구청명, 10분우량 열 추출
df = df.iloc[:,[3,4,5]]
```

```python
# 자료수집 시각, 구청명 별 10분우량의 평균 피봇 테이블 생성
result = df.pivot_table(index=['자료수집 시각', '구청명'], values=['10분우량'], aggfunc=['mean'])
```

```python
# 피봇테이블을 df로 변환
result.reset_index(inplace=True)
result.head(40)
```

```python
# 10분우량 평균 반올림
result['mean'] = result['mean'].round(2)
result
```

```python
result.to_csv('./rain.csv', index=False,encoding='utf-8-sig')
```
