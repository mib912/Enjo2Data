```python
data1 = df[['rent_cnt', 'dust', 'temp', 'road_cnt', 'road_len', 'rain']]
data1
```

### 표준화

#### StandardScaler 이용

```python
# StandardScaler 이용 표준화해서 변환

from sklearn.preprocessing import StandardScaler

# StandardScaler객체 생성
scaler = StandardScaler()
# StandardScaler 로 데이터 셋 변환. fit( ) 과 transform( ) 호출.  
scaler.fit(data1)
data1_scaled = scaler.transform(data1) # numpy ndarry 반환

# numpy ndarry를 DataFrame으로 변환
data1_df_scaled = pd.DataFrame(data=data1_scaled, columns=data1.columns)

data1_df_scaled.head()


# 모든 컬럼 값의 평균이 0에 아주 가까운 값으로,
# 분산은 1에 아주 가까운 값으로 변환
```

```python
data1_df_scaled
```

#### 표준화한 값과 기존 df 합치기

```python
data2 = df
data2
```

```python
# data2의 dust, temp, road_cnt,road_len, rain 열 삭제
data2 = data2.drop(['rent_cnt','dust', 'temp', 'road_cnt', 'road_len', 'rain'], axis=1)
data2
```

```python
# data2와 data1_df_scaled 합치기
data3 = pd.concat([data1_df_scaled,data2], axis=1)
```

```python
data3
```

```python
data3.isnull().sum()
```

#### season 열 원핫 인코딩

```python
df2 = data3.season
season = pd.get_dummies(df2)
season
```

```python
season.columns
season.columns = ['봄','여름','가을','겨울']
season
```

```python
data4 = pd.concat([data3, season], axis=1)
data4
```

```python
data4 = data4.drop('season',axis=1)
```

#### MinMaxScaler 이용

```python
from sklearn.preprocessing import MinMaxScaler

# MinMaxScaler객체 생성
scaler = MinMaxScaler()
# MinMaxScaler 로 데이터 셋 변환. fit() 과 transform() 호출.  
scaler.fit(data1)
data1_scaled = scaler.transform(data1) # numpy ndarry로 반환

# numpy ndarry를 DataFrame으로 변환
data1_df_scaled = pd.DataFrame(data=data1_scaled, columns=data1.columns)
print('feature들의 최소 값')
print(data1_df_scaled.min())
print('\nfeature들의 최대 값')
print(data1_df_scaled.max())
```
