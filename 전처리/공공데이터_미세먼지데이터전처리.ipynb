```python
dust.loc[(dust['미세먼지']>=0)&(dust['미세먼지']<=35),'상태']='보통'
dust.loc[(dust['미세먼지']>=36)&(dust['미세먼지']<= 75),'상태']='나쁨'
dust.loc[dust['미세먼지']>=76,'상태']='매우나쁨'
dust
```

- 16-35(보통) = 보통
- 36-75(나쁨) = 나쁨
- 76이상(매우나쁨) = 매우나쁨
