```python
import pandas as pd
import numpy as np
```

```python
weather=pd.read_csv('./data/미세먼지, 날씨 데이터/2021년일별강수량.csv',encoding='euc-kr',header=7)
weather
```

```python
weather_sample=weather.drop('지점',axis=1)
weather_sample
```

```python
weather_sample.isnull().sum()
# 총 365개중 205 강수량 측정 x -> 0으로 바꿔줄 거임.
```

```python
weather_sample = weather_sample.fillna(0)
weather_sample
```

```python
weather_sample.isnull().sum()
# 결측치 제거 완료
```

강수량 ( 기준은 통계청 기준)
- 0 맑음  = 1
- 3mm 미만 = 2 
- 3~15미만(보통) = 3
- 15이상(강비) = 4
- 30이상(폭우) = 5

```python
weather_sample['날씨']=0
```

```python
weather_sample.loc[weather_sample['강수량(mm)']>=0.0,'날씨']='1'
weather_sample.loc[(weather_sample['강수량(mm)']>0.0)&(weather_sample['강수량(mm)']<3.0),'날씨']='2'
weather_sample.loc[(weather_sample['강수량(mm)']>=3.0)&(weather_sample['강수량(mm)'] < 15.0),'날씨']='3'
weather_sample.loc[(weather_sample['강수량(mm)']>=15.0)&(weather_sample['강수량(mm)'] < 30.0),'날씨']='4'
weather_sample.loc[weather_sample['강수량(mm)']>=30.0,'날씨']='5'
weather_sample
```

```python
weather_sample.날씨.value_counts()
```

```python
pd.to_numeric(weather_sample['강수량(mm)'])
```

```python
weather_sample.to_csv('weather.csv')
```
