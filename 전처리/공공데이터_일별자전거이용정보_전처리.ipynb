```python
data=pd.concat([jan,feb,mar,apr,may,jun,jul,aug,sep,octo,nov,dec])
```

```python
data.head(10)
```

```python
data["성별"]=data['성별'].fillna('N') # 공백 데이터 N으로 변경
```

```python
data["성별"] = data["성별"].str.replace(pat=r'[^\w]', repl=r'', regex=True)
data.head(10) # 특수문자 없애기
```
