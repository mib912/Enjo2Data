```python
def get_title(url):
    html = requests.get(url).content
    soup = BeautifulSoup(html,'html.parser')
    
    ex_titles = soup.find('ul', {'class':'list_df cate_documents'})
    lis = ex_titles.findAll('li')
    
    title_list = [get_title_info(li) for li in lis]
    
    return title_list
```

```python
def get_title_info(li):
    title = li.find('a').text
    link = li.find('a')['href']
    
    return {'TITLE':title, 'LINK':link}
```
