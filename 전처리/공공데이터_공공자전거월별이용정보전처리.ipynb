```python
sep=sep.rename(columns={'이용거리(M)':'이동거리(M)'})
sep=sep.rename(columns={'이용시간(본)':'이용시간(분)'}) # sep column오타수정
```

```python
data=pd.concat([jan,feb,Mar,apr,may,jun,jul,aug,sep,octo,nov,dec])
data.head(10)
```

```python
data['성별'].fillna('N') # 공백 데이터 N으로 변경
```

```python
data["성별"] = data["성별"].str.replace(pat=r'[^\w]', repl=r'', regex=True)
data.head(10) # 특수문자 없애기
```

```python
data['대여구분코드'].unique()
```

```python
data['대여구분코드']=data['대여구분코드'].replace('10분이용권','일일(비회원)')
data['대여구분코드']=data['대여구분코드'].replace('BIL_021','일일(비회원)')
```

```python
data['대여구분코드'].value_counts() # 10분 이용권과 BIL_021을 일일(비회원)으로 옮김
```
